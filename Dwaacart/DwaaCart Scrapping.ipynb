```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import datetime
import time
import csv
```

```python
url = "https://dwaacart.com/prod_details/%E2%80%8EColoplast"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"}
```

```python
page = requests.get(url,headers = headers)
page
```

```python
page.content
```

```python
type(page.content)
```

```python
soup1 = BeautifulSoup(page.content , "html.parser")
print(soup1)
```

```python
soup2 = BeautifulSoup(soup1.prettify() , "html.parser")
```

```python
Name_element = soup2.select_one(".title")
Name = Name_element.text.strip()
Price_element = soup2.select_one(".price")
Price = Price_element.text.strip()[1:4]
Date = datetime.date.today()
print(Name)
print(Price)
print(Date)
```

```python
header = ["Date" , "Title" , "Price" ]
data = [Date , Name , Price]


```

```python
with open(file="Dwaacart3.csv" , mode='w',newline='',encoding="UTF8") as f :
    writer = csv.writer(f)
    
    writer.writerow(header)
    writer.writerow(data)
```
